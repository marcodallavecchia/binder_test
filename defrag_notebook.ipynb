## Importing packages

```python
import skimage.filters
import skimage.io
import matplotlib.pyplot as plt
import pandas as pd
```

## Image import

Here we use a link to an image present in a GitHub repository.

```python
im = skimage.io.imread('https://github.com/guiwitz/PyImageCourse_beginner/raw/master/images/19838_1252_F8_1.tif')
```

```python
plt.imshow(im);
```

## Processing

We smooth the image and then threshold it and label it.

```python
im_gauss = skimage.filters.gaussian(im[:,:,2], sigma=5)
im_th = im_gauss > skimage.filters.threshold_otsu(im_gauss)
im_label = skimage.measure.label(im_th)
```

```python
plt.imshow(im_label);
```

## Detect objects

```python
nuclei = skimage.measure.regionprops_table(im_label, properties=('label', 'area'))
```

```python
nuclei = pd.DataFrame(nuclei)
```
